```python
#create EMPTY Lists:
AN=[] #airline name
DOJ=[] #date of journey
S=[] #source
D=[] #destination
R=[] #route
DT=[] #departure time
AT=[] #arrival time
Duration=[] #duration
TS=[] #Total stops 
Price=[] #price
```

```python
start=0
end=138
for page in range(start,end):#for loop for scrapping 3 page
    an_tags=driver.find_elements_by_xpath("//div[@class='css-1dimx8f']") #locating web elements for the airline name title
    for i in an_tags:
        AN.append(i.text.replace('\n',""))

    doj_tags=driver.find_elements_by_xpath("//div[@class='css-5nu86q']") #locating web elements for date of journey
    for i in doj_tags:
        DOJ.append(i.text.replace('\n','').split()[2:4])

    s_tags=driver.find_elements_by_xpath("//div[@class='css-5nu86q']") #locating web elements for departure city
    for i in s_tags:
        S.append(i.text.split()[0])

    d_tags=driver.find_elements_by_xpath("//div[@class='css-yyi517']") #locating web elements for destination city
    for i in d_tags:
        D.append(i.text)
D=D[0::2]

    dt_tags=driver.find_elements_by_xpath("//div[@data-testid='flight_card_segment_departure_time']") #locating web elements for departure time
    for i in dt_tags:
        DT.append(i.text)

    at_tags=driver.find_elements_by_xpath("//div[@data-testid='flight_card_segment_destination_time']") #locating web elements for arrival time
    for i in at_tags:
        AT.append(i.text)

    dur_tags=driver.find_elements_by_xpath("//div[@data-testid='flight_card_segment_duration']") #locating web elements for flight duration
    for i in dur_tags:
        Duration.append(i.text)

    ts_tags=driver.find_elements_by_xpath("//div[@data-testid='flight_card_segment_stops']") #locating web elements for total stops
    for i in ts_tags:
        TS.append(i.text)
TS
```
